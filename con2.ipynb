```python
import pandas as pd
df = pd.read_csv("Sales.csv")
```

```python
import pandas as pd

unique_pizza_ids = df['pizza_name_id'].unique()

dataframes = {}

for pizza_id in unique_pizza_ids:
    df_filtered = df[df['pizza_name_id'] == pizza_id][['week_number', 'quantity', 'order_date']]
    dataframes[pizza_id] = df_filtered
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error

results = {}

for j in range(0, 91):
    example_pizza_id = unique_pizza_ids[j]
    df = dataframes[example_pizza_id]
    
    # Group by 'week_number' and sum 'quantity'
    df_weekly = df.groupby('week_number').sum().reset_index()
    df_weekly.rename(columns={'week_number': 'ds', 'quantity': 'y'}, inplace=True)
    
    # Create weekly date range starting from min order date
    df_weekly['ds'] = pd.date_range(start=df['order_date'].min(), periods=len(df_weekly), freq='W')

    # Split the data into training (85%) and test (15%)
    split_point = int(len(df_weekly) * 0.85)
    train_df = df_weekly[:split_point]
    test_df = df_weekly[split_point:]

    # Fit ARIMA model on the training data (p, d, q need to be chosen based on the data)
    arima_model = ARIMA(train_df['y'], order=(1,1,1))  # Example ARIMA(p=1, d=1, q=1)
    arima_fit = arima_model.fit()

    # Forecast for the length of the test set
    forecast_test = arima_fit.forecast(steps=len(test_df))
    
    # Calculate MAPE for the test data
    mape = mean_absolute_percentage_error(test_df['y'], forecast_test)

    # Forecast for the next 'len(test_df)' weeks after the test set
    future_forecast = arima_fit.forecast(steps=len(test_df) + 1)
    
    # Get the first prediction (next week after the test set)
    next_week_forecast = round(future_forecast.iloc[0])

    # Store the results in the dictionary
    results[f'unique_pizza_ids[{j}]'] = {
        'Pizza_id': unique_pizza_ids[j],
        'MAPE': mape,
        'Next_Week_Prediction': next_week_forecast
    }

# Results can be accessed via the 'results' dictionary
```

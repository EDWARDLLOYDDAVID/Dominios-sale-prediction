```python
import pandas as pd

df1 = pd.read_csv('Pizza_Sale - pizza_sales.csv')
df2 = pd.read_csv('Pizza_ingredients - Pizza_ingredients.csv')
```

```python
df1 = df1.drop(['pizza_id','order_id','pizza_category'],axis = 1)
df1
```

```python
df2.isna().sum()
```

```python
import pandas as pd

# Fill NaN values in column3 with the product of column1 and column2
df1['total_price'] = df1.apply(lambda row: row['quantity'] * row['unit_price'] if pd.isna(row['total_price']) else row['total_price'], axis=1)

print(df1)
```

```python
df1.isna().sum()
```

```python
df2.isna().sum()
```

```python

pizza_dict = df2.set_index('pizza_name_id')['pizza_name'].to_dict()

df1['pizza_name'] = df1.apply(lambda row: pizza_dict[row['pizza_name_id']] if pd.isna(row['pizza_name']) else row['pizza_name'], axis=1)

print(df1)
```

```python
ingredient_dict = df2.set_index('pizza_name')['pizza_ingredients'].to_dict()

# Fill missing values in df1 using the dictionary
df1['pizza_ingredients'] = df1.apply(lambda row: ingredient_dict[row['pizza_name']] if pd.isna(row['pizza_ingredients']) else row['pizza_ingredients'], axis=1)

print(df1)
```

```python
ingredient_dict = df2.set_index('pizza_name')['pizza_name_id'].to_dict()

# Fill missing values in df1 using the dictionary
df1['pizza_name_id'] = df1.apply(lambda row: ingredient_dict[row['pizza_name']] if pd.isna(row['pizza_name_id']) else row['pizza_name_id'], axis=1)

print(df1)
```
